# Age Prediction based on Knee MRIs

This notebook trains a CNN to predict the age of a person based on their knee MRI. It first applies a segmentation on the bone to reduce the complexity of the problem and then continues to run a regression on the age.

## 1 Imports

```python
import numpy as np
import dask.array as da
from ipywidgets import FloatProgress
from IPython.display import display
from utils import *
%matplotlib inline
import matplotlib.pyplot as plt
from glob import glob
```

## 2 Data Preprocessing

Load the data while applying cropping and resizing

```python
x_jop, y_jop = load_data_age('data/Jopp_Study/MHD/*.mhd', crop=(128,128,256), 
                             size=(224,224), smart_crop=True)
x_epi, y_epi = load_data_age('data/Epi_Study/MHD/*.mhd', crop=(120,120,560), 
                             size=(224,224), smart_crop=False)
```

The first and last slices on the z-axis show no bone matter and are therefor left empty after the segmentation is applied. I remove them here to get them out the way and also bring both data sources to the same z-axis resolution of 18 slices.

Note: The segmentation network is trained on all slices.

```python
x_jop = x_jop[:, 4:-2]
x_epi = x_epi[:, 12:-11]
```

Create Train, Validation and Test set and convert data to 2D. 

- 2124 (118x18) samples in the training set
- 234 (13x18) samples in the validation set
- 252 (14x18) samples in the test set

```python
x_tr = np.concatenate((to_2d(x_epi[:54]), to_2d(x_jop[:64])))
y_tr = np.concatenate((lengthen(y_epi[:54], 18), lengthen(y_jop[:64], 18)))
x_va = np.concatenate((to_2d(x_epi[54:59]), to_2d(x_jop[64:72])))
y_va = np.concatenate((lengthen(y_epi[54:59], 18), lengthen(y_jop[64:72], 18)))
x_te = np.concatenate((to_2d(x_epi[59:]), to_2d(x_jop[72:])))
y_te = np.concatenate((lengthen(y_epi[59:], 18), lengthen(y_jop[72:], 18)))
```

Normalize data based on mean and standard deviation from segmentation training set.

```python
x_tr = (x_tr - 140) / 170.
x_va = (x_va - 140) / 170.
x_te = (x_te - 140) / 170.
```

Load the pretrained segmentation model and predict maps for the entire data

```python
model = UNet(x_tr.shape[1:], 1, 32, 4, 1, 'elu', upconv=False)
model.compile(optimizer=Adam(lr=0.001, decay=0.001), loss=f1_loss)
model.load_weights('weights/seg2d_9786_std_aug_deconv.h5')
```

```python
x_tr_mask = model.predict(x_tr, batch_size=16, verbose=1)
x_va_mask = model.predict(x_va, batch_size=16, verbose=1)
x_te_mask = model.predict(x_te, batch_size=16, verbose=1)
```

Revert the normalization, so all values are positive

```python
x_tr = x_tr * 170 + 140
x_va = x_va * 170 + 140
x_te = x_te * 170 + 140
```

Check a few samples of the data if everything seems right

```python
show_samples(x_tr, x_tr_mask, 20)
```

Samples look right, although they are definitely not perfect. It may be because the segmentation was trained on N4 Bias Field corrected data, while this data is **not**.

Let's take a look at the labels.

```python
print('Oldest candidate:', round(y_tr.max(),1))
print('Youngest candidate:', round(y_tr.min(),1))
print('Age range:', round(y_tr.max()-y_tr.min(),1))
print('Average age:', round(y_tr.mean(),1))
print('Baseline:', round(abs(y_tr.mean()-y_tr).mean(),1))
```

The ages vary between 14.4 and 20.9 years creating a range of 6.5 years. The mean is located at 17.4 years. If we were to predict the mean age for every sample we would get a mean absolute error of 1.2 years. This is the value to beat.

Apply the segmentation map on the raw input data to remove everything that is not bone.

```python
x_tr = x_tr * x_tr_mask
x_va = x_va * x_va_mask
x_te = x_te * x_te_mask
```

Normalize the data again for training

```python
x_tr_mean = x_tr.mean()
x_tr_std = x_tr.std()
x_tr = (x_tr - x_tr_mean) / x_tr_std
x_va = (x_va - x_tr_mean) / x_tr_std
x_te = (x_te - x_tr_mean) / x_tr_std
print(x_tr_mean, x_tr_std)
```

Augment the training data using horizontal flips and horizontal shift at 20px. This will virtually increase the training data by a factor of 6.

```python
x_tr, y_tr = augment(x_tr, y_tr, h_flip=True, h_shift=[-20, 20])
```

Let's look at the samples one more time with the segmentation mask applied.

```python
show_samples(x_tr, x_tr, 20)
```

```python
print(x_tr.shape, y_tr.shape)
```

## 3 Training

The segmentation model from above uses a encoder-decoder architecture, which will first reduce the spatial resolution of inputs similar to common classification networks. Then on the second half expand the resolution again to create a full segmentation map.

How about we use the first half of that architecture as a pre-trained model for the age prediction? Delete the decoding layers and append 2 new ones for the regression output.

```python
model = UNet(x_tr.shape[1:], 1, 32, 4, 1, 'elu', upconv=False)
model.load_weights('weights/seg2d_9786_std_aug_deconv.h5')
```

```python
del model.layers[-17:]
```

```python
i = Input(shape=x_tr.shape[1:])
x = model.layers[-1].output
x = GlobalAveragePooling2D()(x)
x = Dense(1, activation='linear')(x)
model = Model(inputs=model.input, outputs=x)
```

Let's take a look what that architecture looks like.

```python
model.summary()
```

Compile the model with MSE as a loss function. Let's add a MAE metric to give an intuitive judgement of the performance.

```python
model.compile(loss='mse', optimizer=Adam(lr=0.001, decay=0.01), metrics=['mae'])
```

Save the best performing state of the model based on the validation data

```python
mc = ModelCheckpoint('weights/age_pred_weights.h5', save_weights_only=True, save_best_only=True)
```

Start the training.

```python
model.fit(x_tr, y_tr, validation_data=(x_va, y_va), epochs=20, batch_size=64, callbacks=[mc])
```

```python
model.compile(loss='mse', optimizer=Adam(lr=0.0001, decay=0.01), metrics=['mae'])
mc = ModelCheckpoint('weights/age_pred_weights_1.h5', save_weights_only=True, save_best_only=True)
model.fit(x_tr, y_tr, validation_data=(x_va, y_va), epochs=10, batch_size=64, callbacks=[mc])
```

The training looks pretty stable and the best validation result does not appear to be an outlier. There are a few throwbacks in performance but judging the small validation size this seems reasonable.

Load the best performing model.

```python
model.load_weights('weights/age_pred_weights.h5')
```

Compare the validation results with those in the training output log

```python
evl = model.evaluate(x_va, y_va, verbose=0)
print('MSE: ' + str(round(evl[0],2)) + ' years')
print('MAE: ' + str(round(evl[1],2)) + ' years')
```

Predict the entire training and validation data

```python
age_vec_pred_tr = model.predict(x_tr, verbose=1).sum(axis=-1)
age_vec_pred_va = model.predict(x_va, verbose=1).sum(axis=-1)
```

Since there will be 1 prediction for every slice let's reshape the 1D array to a 2D array where the second axis shows each of the 18 predictions per 3D input image

```python
num_slc = 18 # number of slices
age_vec_pred_tr = np.reshape(age_vec_pred_tr, (len(age_vec_pred_tr) // num_slc, num_slc))
age_vec_pred_va = np.reshape(age_vec_pred_va, (len(age_vec_pred_va) // num_slc, num_slc))
age_tr = shorten(y_tr, num_slc)
age_va = shorten(y_va, num_slc)
```

We now have 18 age predictions per 3D image. Since slices in the middle will most likely show higher accuracy because they contain more information, we could weight these 18 outputs accordingly. This would create a bell shape for example.

We could also train another tree model to figure out the best accuracy representation in each vector.

```python
from sklearn.ensemble import GradientBoostingRegressor, ExtraTreesRegressor
from sklearn.linear_model import LinearRegression, LogisticRegression
reg = ExtraTreesRegressor(max_depth=3)
reg.fit(age_vec_pred_tr[:2124], age_tr[:2124])
```

Let's look at the number of training and validation

```python
age_pred_tr = reg.predict(age_vec_pred_tr)
abs(age_pred_tr - age_tr).mean()
```

```python
age_pred_va = reg.predict(age_vec_pred_va)
abs(age_pred_va - age_va).mean()
```

## 6 Final Verification

At this point I allow myself a run on the test set to see if any of this makes sense. We have not used the test set in any way, except for normalizing it on the training data. We have no information how the pipeline above performs on the test set.

```python
age_vec_pred_te = model.predict(x_te, verbose=1).sum(axis=-1)
age_vec_pred_te = np.reshape(age_vec_pred_te, (len(age_vec_pred_te) // num_slc, num_slc))
age_te = shorten(y_te, num_slc)
age_pred_te = reg.predict(age_vec_pred_te)

te_mae = abs(age_pred_te - age_te).mean()
print(' ')
print(' ')
print('The mean absolute error across ' + str(len(age_pred_te)) + ' samples is ' + str(round(te_mae, 3)) + ' years.')
```

### Possible defects in this pipeline

- Since the validation and test sets are very small, the results may be inaccurate.
- I did not shuffle the data before seperating them into the 3 splits, because the sources contain multiple recordings of the same person. Shuffling them could have meant that a person from the training set also shows up in the validation or test set. I manually checked the age distribution for validation and test - they looked fair. I assumed that there is no bias in the order of the data, which may be optimistic.
